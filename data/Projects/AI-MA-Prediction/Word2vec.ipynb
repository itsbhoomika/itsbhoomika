```python
import re
import ast

def fix_broken_ner(text):
    if not isinstance(text, str):
        return text
    text_fixed = re.sub(r'\]\s*\[', ',', text)     # Fix broken small list joins
    text_fixed = re.sub(r',\s*,+', ',', text_fixed) # Fix double commas
    text_fixed = re.sub(r'\[\s*,\s*', '[', text_fixed)  # Fix opening bracket followed by comma
    return ast.literal_eval(text_fixed)

# Apply to fix NER_combined
grouped_df['NER_combined'] = grouped_df['NER_combined'].apply(fix_broken_ner)
```

```python
grouped_df.head()
```

```python
corpus = []
for ner_list in grouped_df['NER_combined']:
    tokens = [entity.replace(' ', '_') for entity, label in ner_list]
    tokens = list(dict.fromkeys(tokens))  # Deduplicate inside one RIC
    corpus.append(tokens)
```

```python
from gensim.models import Word2Vec

w2v_model = Word2Vec(
    sentences=corpus,
    vector_size=100,
    window=2,
    min_count=1,
    workers=4,
    sg=1,
    sample=1e-3
)
```

```python
import numpy as np

def get_ric_vector(ner_list):
    tokens = [entity.replace(' ', '_') for entity, label in ner_list]
    tokens = list(dict.fromkeys(tokens))  # deduplicate
    vectors = []
    for token in tokens:
        if token in w2v_model.wv:
            vectors.append(w2v_model.wv[token])
    if vectors:
        return np.mean(vectors, axis=0)  # Average embedding
    else:
        return np.zeros(w2v_model.vector_size)

grouped_df['RIC_Vector'] = grouped_df['NER_combined'].apply(get_ric_vector)
```

```python
grouped_df.head()
```

```python
import pandas as pd

# Step 1: Expand RIC_Vector into separate columns
ric_vector_df = grouped_df['RIC_Vector'].apply(lambda x: pd.Series(x))

# Step 2: Rename columns as word2vec_0, word2vec_1, ..., word2vec_99
ric_vector_df.columns = [f'word2vec_{i}' for i in range(ric_vector_df.shape[1])]

# Step 3: Concatenate back with RIC for clarity
final_word2vec_df = pd.concat([grouped_df[['RIC']], ric_vector_df], axis=1)
```

# Feature Selection with RFE

```python
# Step 1: Drop target and unnecessary columns
X = merged.drop(columns=["Instrument", "AD", "AD-30", "Label"])
y = merged["Label"]

# Step 2: Keep only numerical columns
X = X.select_dtypes(include=['number'])  # Keep only int/float columns

# Step 3: Scaling (optional for Random Forest but good for consistency)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Feature selection with RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE

model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight="balanced")
rfe = RFE(estimator=model, n_features_to_select=50, step=10, verbose=1)
X_selected = rfe.fit_transform(X_scaled, y)

# Step 5: Get feature names
selected_features = X.columns[rfe.get_support()]
X_selected_df = pd.DataFrame(X_selected, columns=selected_features)
```
